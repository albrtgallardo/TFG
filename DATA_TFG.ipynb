### df del promedio de los valores del QAOA

```python
df_QAOA_LN = df_QAOA_LN_1.copy()
```

```python
#como no variamos la tolerancia y el learning rate, mejor eliminar esas columnas
df_QAOA_LN = df_QAOA_LN.drop(columns=['Toler', 'Learning rate', 'Tiempo ejecucion','Primer minimo','Δ True Min'])
```

```python
avg_time_QAOA_LN = (df_QAOA_LN_1['Tiempo ejecucion'] + df_QAOA_LN_2['Tiempo ejecucion'] + df_QAOA_LN_3['Tiempo ejecucion'])/3
std_time_QAOA_LN = pd.concat([df_QAOA_LN_1['Tiempo ejecucion'], df_QAOA_LN_2['Tiempo ejecucion'], df_QAOA_LN_3['Tiempo ejecucion']], axis=1).std(axis=1)


df_QAOA_LN['Promedio tiempo'] = avg_time_QAOA_LN
df_QAOA_LN['Tiempo std'] = std_time_QAOA_LN
df_QAOA_LN['CV Tiempo'] = (std_time_QAOA_LN/avg_time_QAOA_LN)*100
```

```python
avg_min_QAOA_LN = (df_QAOA_LN_1['Primer minimo'] + df_QAOA_LN_2['Primer minimo'] + df_QAOA_LN_3['Primer minimo'])/3
std_min_QAOA_LN = pd.concat([df_QAOA_LN_1['Primer minimo'], df_QAOA_LN_2['Primer minimo'], df_QAOA_LN_3['Primer minimo']], axis=1).std(axis=1)


df_QAOA_LN['Minimo promedio'] = avg_min_QAOA_LN
df_QAOA_LN['Minimo std'] = std_min_QAOA_LN
df_QAOA_LN['CV Minimo'] = (std_min_QAOA_LN/avg_min_QAOA_LN)*100
```

```python
avg_tmin_QAOA_LN = (df_QAOA_LN_1['Δ True Min'] + df_QAOA_LN_2['Δ True Min'] + df_QAOA_LN_3['Δ True Min'])/3
std_tmin_QAOA_LN = pd.concat([df_QAOA_LN_1['Δ True Min'], df_QAOA_LN_2['Δ True Min'], df_QAOA_LN_3['Δ True Min']], axis=1).std(axis=1)


df_QAOA_LN['Δ True Min avg'] = avg_tmin_QAOA_LN
df_QAOA_LN['Δ True Min std'] = std_tmin_QAOA_LN
df_QAOA_LN['CV Δ True Min'] = (std_tmin_QAOA_LN/avg_tmin_QAOA_LN)*100
```

```python
display(df_QAOA_LN)
```

### df de todas la ejecuciones

```python
df_QAOA_LN_total = pd.concat([df_QAOA_LN_1, df_QAOA_LN_2, df_QAOA_LN_3], ignore_index=True)
df_QAOA_LN_total = df_QAOA_LN_total.drop(columns=['Toler', 'Learning rate'])

corr_matrix_LN = df_QAOA_LN_total.corr()
display(df_QAOA_LN_total)
```
